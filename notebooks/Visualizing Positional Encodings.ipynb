# Visualizing Positional Encodings

## Imports

```python
import sys
import os

root_path = os.path.abspath(os.path.join('..'))
if root_path not in sys.path:
    sys.path.append(root_path)

import matplotlib.pyplot as plt

import src.models.positional_encoding as pe
```

## Plot positional encodings

```python
def plot_positional_encodings():
    max_length = 2048
    d_model = 512
    positional_encodings = pe.positional_encoding(max_length, d_model)

    plt.pcolormesh(positional_encodings, cmap='RdBu')
    plt.xlabel('Embedding Dimension, $i$')
    plt.ylabel('Token Position, $pos$')
    plt.ylim((max_length, 0))
    plt.colorbar()
    plt.title('Positional Encoding')
    plt.show()
```

```python
plot_positional_encodings()
```
